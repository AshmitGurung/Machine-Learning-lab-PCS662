```python

import numpy as np #make sure to run this cell before the other cell to ensure import
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

```

```python
x=np.linspace(0,3*np.pi,100)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python


x_data=np.linspace(0,3*np.pi,200)
y_data=np.sin(x_data)
plt.plot(x,y,c='g')
plt.scatter(x_data,y_data)
```

```python
y_data_noise=y_data+np.random.normal(0,0.3,200)
plt.plot(x,y,c='g')
plt.scatter(x_data,y_data_noise,c='r')
```

```python
xTx = np.dot(x_data.reshape(x_data.shape[0], 1).T, x_data.reshape(x_data.shape[0], 1))
xTx_inv = np.linalg.inv(xTx)
# Calculate xTy similar to how xTx is calculated, using x_data instead of x
xTy = np.dot(x_data.reshape(x_data.shape[0], 1).T, y_data_noise.reshape(y_data_noise.shape[0],1))
w = np.dot(xTx_inv, xTy)
print(w)
```

```python
degrees=[1,4,7]
plt.figure(figsize=(10,6))
plt.plot(x,y,c='g',label='True Function')
plt.scatter(x_data,y_data_noise,c='r',label='Noisy Data')
for degree in degrees:
    x_data_reshaped=x_data.reshape(-1,1)
    poly=PolynomialFeatures(degree=degree)

    X_poly=poly.fit_transform(x_data_reshaped)
    model=LinearRegression()
    model.fit(X_poly,y_data_noise)

    x_reshaped=x.reshape(-1,1)
    X_poly_pred=poly.transform(x_reshaped)
    y_pred=model.predict(X_poly_pred)

    plt.plot(x,y_pred,label=f'Degree {degree}')

plt.legend()
plt.show()
```
